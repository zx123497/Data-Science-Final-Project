```python
from datascience import *
import numpy as np
shelter_url = './animal_shelter.csv'
np.set_printoptions(threshold=50)
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
matplotlib.rcParams['font.family'] = ['Heiti TC']
```

```python
animal_shelter_raw = Table.read_table(shelter_url)
animal_shelter_raw
```

```python
shelter_info = animal_shelter_raw.select("animal_id","animal_kind","animal_Variety","shelter_name","shelter_address")
shelter_info
```

```python
place = shelter_info.apply(lambda inp: inp[:3], "shelter_name")
shelter_info = shelter_info.with_column("place",place)
shelter_info
```

```python
place_group = shelter_info.group("place").sort("count",True)
```

```python
dog_cat_place = shelter_info.pivot("animal_kind","place").drop("其他").sort("place")
```

```python
annual_animal = Table.read_table('./annual_animal.csv')
annual_animal.show()
```

```python
animal_110 = annual_animal.where("rpt_year",are.equal_to(110))
animal_110.show()
```

```python
animal_110 = animal_110.relabeled("calcu_dog_num","狗").relabeled("calcu_cat_num","貓")
cat_dog_110 = animal_110.select("rpt_country",'狗',"貓")
```

## 貓狗總數統計

```python
cat_dog_110.sort("rpt_country").barh("rpt_country")
```

## 未領養貓狗總數＆未領養動物總數統計

```python
dog_cat_place.barh("place")
place_group.barh("place")
```

### 觀察＆推論
- 上面兩個個表可以看出：
    1. 台灣狗比貓多
    2. 都市(六都)動物比鄉下多
    3. 動物總計與開放認領養動物數量有正相關
    4. 澎湖動物總數少，開放認領影數量卻很多
    5. 六都中，台中開放認領＆家犬家貓總數差異極大
    
- 推測：
    1. 都市的認領養機構較多且完善，所以有較多的狗狗貓貓開放認領
    2. 因為都市人多，較多收容所＋養寵物的人，（生存容易？）
    3. 澎湖人不喜歡養寵物，導致收容所的動物沒有人要領養
    4. 是否因為貓咪比較怕人，未統計到的黑數比狗狗多？
    5. 台中人喜歡養寵物？ 因為綠園道那裡假日超多人遛狗

```python
all_110 = cat_dog_110.apply(lambda x, y: x + y, "狗", "貓")
all_110
```

```python
cat_dog_110 = cat_dog_110.with_column("animal_sum", all_110)
cat_dog_110
```

```python
shelter_110 = animal_shelter_raw.where("animal_opendate",are.not_containing("2022"))
shelter_110 = shelter_110.where("animal_status", are.equal_to("OPEN"))
```

```python
shelter_110 = shelter_110.select("animal_kind","animal_opendate","shelter_name")
shelter_110 = shelter_110.with_column("place", shelter_110.apply(lambda x: x[:3],"shelter_name"))
shelter_110.sort("animal_opendate")
```

```python
shelter_place_110 = shelter_110.group("place")
```

```python
cat_dog_shelter_110 = shelter_110.pivot("animal_kind","place").drop("其他")
cat_dog_shelter_110 = cat_dog_shelter_110.join("place",shelter_place_110)
cat_dog_shelter_110 = cat_dog_shelter_110.relabeled("狗","110_shelter_dog").relabeled("貓","110_shelter_cat").relabeled("count","110_shelter_sum")
```

```python
cat_dog_shelter_110
```

```python
animal_110 = animal_110.relabeled("rpt_country","place")
```

```python
animal_110 = animal_110.join("place",cat_dog_shelter_110)
```

```python
animal_110 = animal_110.drop("els_note").drop("stray_dog_num")
animal_110
```

```python
animal_110 = animal_110.with_column("total_animal",animal_110.apply(lambda x, y: x+y, "貓", "狗"))
animal_110 = animal_110.with_column("shelter percentage", animal_110.column("110_shelter_sum")/animal_110.column("total_animal"))
```

```python
animal_110 = animal_110.set_format('shelter percentage', PercentFormatter).sort("total_animal",True)
```

```python
animal_110.show()
```

## 未領養動物比例

```python
animal_110.select("place","shelter percentage").sort("shelter percentage",True).barh("place")
```
